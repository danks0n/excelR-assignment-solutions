```python
df = pd.read_csv("ToyotaCorolla.csv", encoding='latin1')
```

```python
df.info()
```

```python
df2=pd.concat([df.iloc[:,2:4],df.iloc[:,6:7],df.iloc[:,8:9],df.iloc[:,12:14],df.iloc[:,15:18]],axis=1)
```

```python
df2.head()
```

```python
df3 = df2.rename({'Age_08_04' : 'Age', 'cc' : 'CC', 'Quarterly_Tax' : 'QT'}, axis = 1)
```

```python
df3[df3.duplicated()]
```

```python
df4[df4.duplicated()]
```

```python
df4.corr()
```

```python
df4.info()
```

```python
x = df4.iloc[:,1::]
```

```python
x.head()
```

```python
y = df4['Price']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0 )
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
#Now that we have created our model, we will use x_test to predict the price.
```

```python
y_pred = model.predict(x_test)
```
